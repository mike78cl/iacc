# Proyecto BikeCity

## Objetivo
Analizar el uso de estaciones de bicicletas compartidas para apoyar decisiones operativas relacionadas con disponibilidad, reclamos y redistribución de bicicletas.

## Preguntas de investigación
1. ¿Qué estaciones presentan mayor demanda?
2. ¿Dónde existen más reclamos?
3. ¿Qué estaciones presentan menor disponibilidad?
4. ¿Qué zonas requieren redistribución de bicicletas?

## Metodología
Se cargó un dataset con 300 registros, se revisó la calidad de los datos, se calcularon métricas operativas y se construyó un dashboard 2x2 para responder las preguntas del análisis.

## Resultados
El dashboard permite identificar estaciones con alta demanda, comunas con mayor cantidad de reclamos, estaciones con baja disponibilidad y zonas con mayor ocupación.

## Decisiones recomendadas
Redistribuir bicicletas, reforzar estaciones críticas, aumentar capacidad en comunas con alta ocupación y monitorear alertas operativas.

## Reflexión ética
El análisis debe utilizarse para mejorar el servicio sin afectar injustamente a comunas o usuarios. Las decisiones deben considerar contexto, demanda real y limitaciones de los datos.

```python
# ============================================================
# Proyecto BikeCity - Análisis de bicicletas compartidas
# ============================================================

# 1. Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from joblib import dump

# 2. Cargar dataset
df = pd.read_excel("bicicletas_compartidas_300_registros.xlsx")

# 3. Revisar primeras filas
print(df.head())

# 4. Revisar estructura general
print(df.info())

# 5. Evaluar calidad de datos
print("Valores nulos por columna:")
print(df.isnull().sum())

print("Registros duplicados:")
print(df.duplicated().sum())

# 6. Crear nuevas métricas
df["porcentaje_ocupacion"] = (
    (df["capacidad_total"] - df["bicicletas_disponibles"]) /
    df["capacidad_total"]
) * 100

df["saldo_operativo"] = df["devoluciones_dia"] - df["prestamos_dia"]

df["nivel_alerta"] = np.where(
    df["bicicletas_disponibles"] <= 3,
    "Alta demanda",
    np.where(
        df["porcentaje_ocupacion"] >= 90,
        "Riesgo saturación",
        "Normal"
    )
)

# 7. Indicadores generales
promedio_prestamos = np.mean(df["prestamos_dia"])
desviacion_prestamos = np.std(df["prestamos_dia"], ddof=1)

intervalo_confianza = stats.t.interval(
    confidence=0.95,
    df=len(df["prestamos_dia"]) - 1,
    loc=promedio_prestamos,
    scale=stats.sem(df["prestamos_dia"])
)

resultados = {
    "promedio_prestamos": promedio_prestamos,
    "desviacion_prestamos": desviacion_prestamos,
    "ic95_prestamos": intervalo_confianza
}

dump(resultados, "resultados_bikecity.joblib")

print("Resultados generales:")
print(resultados)

# 8. Agrupaciones para responder preguntas
demanda_estacion = df.groupby("estacion")["prestamos_dia"].mean().sort_values(ascending=False)

reclamos_comuna = df.groupby("comuna")["reclamos"].sum().sort_values(ascending=False)

disponibilidad_estacion = df.groupby("estacion")["bicicletas_disponibles"].mean().sort_values()

ocupacion_comuna = df.groupby("comuna")["porcentaje_ocupacion"].mean().sort_values(ascending=False)

print("Estaciones con mayor demanda:")
print(demanda_estacion.head())

print("Comunas con más reclamos:")
print(reclamos_comuna.head())

print("Estaciones con menor disponibilidad:")
print(disponibilidad_estacion.head())

print("Comunas con mayor ocupación:")
print(ocupacion_comuna.head())

# 9. Exportar reportes para distintas áreas

# Reporte para Operaciones
reporte_operaciones = df[
    [
        "fecha",
        "estacion",
        "comuna",
        "bicicletas_disponibles",
        "capacidad_total",
        "prestamos_dia",
        "devoluciones_dia",
        "saldo_operativo",
        "nivel_alerta"
    ]
]

reporte_operaciones.to_excel(
    "reporte_operaciones_bikecity.xlsx",
    index=False
)

# Reporte para Atención al Cliente
reporte_clientes = df[
    [
        "fecha",
        "comuna",
        "estacion",
        "reclamos",
        "bicicletas_disponibles",
        "nivel_alerta"
    ]
]

reporte_clientes.to_excel(
    "reporte_atencion_clientes_bikecity.xlsx",
    index=False
)

# 10. Dashboard 2x2 con respuestas a las preguntas planteadas

# ============================================================
# Dashboard 2x2 con distintos tipos de gráficos
# ============================================================

plt.figure(figsize=(15, 10))

# 1. Gráfico de barras
# Pregunta: ¿Qué estaciones presentan mayor demanda?
plt.subplot(2, 2, 1)
demanda_estacion.head(5).plot(kind="bar")
plt.title("1. Estaciones con mayor demanda")
plt.xlabel("Estación")
plt.ylabel("Préstamos promedio")
plt.xticks(rotation=45)

# 2. Gráfico circular
# Pregunta: ¿Dónde existen más reclamos?
plt.subplot(2, 2, 2)
reclamos_comuna.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)
plt.title("2. Distribución de reclamos por comuna")
plt.ylabel("")

# 3. Gráfico de líneas
# Pregunta: ¿Cómo evoluciona la demanda en el tiempo?
demanda_fecha = df.groupby("fecha")["prestamos_dia"].sum()

plt.subplot(2, 2, 3)
demanda_fecha.plot(kind="line", marker="o")
plt.title("3. Evolución diaria de préstamos")
plt.xlabel("Fecha")
plt.ylabel("Total de préstamos")
plt.xticks(rotation=45)

# 4. Gráfico de dispersión
# Pregunta: ¿Existe relación entre disponibilidad y reclamos?
plt.subplot(2, 2, 4)
plt.scatter(
    df["bicicletas_disponibles"],
    df["reclamos"]
)
plt.title("4. Relación entre disponibilidad y reclamos")
plt.xlabel("Bicicletas disponibles")
plt.ylabel("Reclamos")

plt.tight_layout()
plt.savefig("dashboard_bikecity_2x2.png", dpi=300)
plt.show()

# 11. Decisiones sugeridas según resultados

print("Decisiones sugeridas:")
print("- Redistribuir bicicletas hacia estaciones con mayor demanda.")
print("- Reforzar estaciones con baja disponibilidad promedio.")
print("- Revisar comunas con mayor cantidad de reclamos.")
print("- Aumentar capacidad en zonas con ocupación promedio elevada.")
print("- Monitorear estaciones con alerta de alta demanda o saturación.")
```
