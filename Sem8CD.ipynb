```python
# ============================================================
# PROYECTO BIKECITY
# ============================================================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os


from funciones_estadisticas import calcular_indicadores, evaluar_calidad_datos
from funciones_reportes import crear_reporte_operaciones, crear_reporte_clientes

# ============================================================
# 1. CARGAR DATASET
# ============================================================

df = pd.read_excel("bicicletas_compartidas_300_registros.xlsx")

print("Primeras filas del dataset:")
display(df.head())


# ============================================================
# 2. REVISAR ESTRUCTURA DEL DATASET
# ============================================================

print("Información general:")
df.info()

print("Estadísticas descriptivas:")
display(df.describe())


# ============================================================
# 3. EVALUAR CALIDAD DE DATOS
# ============================================================

calidad = evaluar_calidad_datos(df)

print("Valores nulos por columna:")
print(calidad["valores_nulos"])

print("\nRegistros duplicados:")
print(calidad["duplicados"])

print("\nTipos de datos:")
print(calidad["tipos_datos"])


# ============================================================
# 4. CREAR VARIABLES PARA EL ANÁLISIS
# ============================================================

# Porcentaje de ocupación de cada estación
df["porcentaje_ocupacion"] = (
    (df["capacidad_total"] - df["bicicletas_disponibles"])
    / df["capacidad_total"]
) * 100

# Diferencia entre devoluciones y préstamos
df["saldo_operativo"] = (
    df["devoluciones_dia"] - df["prestamos_dia"]
)

# Clasificación simple del estado de la estación
df["nivel_alerta"] = np.where(
    df["bicicletas_disponibles"] <= 3,
    "Alta demanda",
    np.where(
        df["porcentaje_ocupacion"] >= 90,
        "Riesgo saturación",
        "Normal"
    )
)

print("Dataset con nuevas variables:")
display(df.head())


# ============================================================
# 5. CALCULAR INDICADORES ESTADÍSTICOS
# ============================================================

# Se usa una función externa compartida por otro integrante
indicadores = calcular_indicadores(
    df["prestamos_dia"],
    "indicadores_prestamos_bikecity.joblib"
)

print("Indicadores estadísticos de préstamos diarios:")
print(indicadores)


# ============================================================
# 6. RESPONDER PREGUNTAS DE INVESTIGACIÓN
# ============================================================

# Pregunta 1:
# ¿Qué estaciones presentan mayor demanda?
demanda_estacion = (
    df.groupby("estacion")["prestamos_dia"]
    .mean()
    .sort_values(ascending=False)
)

# Pregunta 2:
# ¿Qué comunas presentan más reclamos?
reclamos_comuna = (
    df.groupby("comuna")["reclamos"]
    .sum()
    .sort_values(ascending=False)
)

# Pregunta 3:
# ¿Cómo evoluciona la demanda a lo largo del tiempo?
demanda_fecha = (
    df.groupby("fecha")["prestamos_dia"]
    .sum()
)

# Pregunta 4:
# ¿Existe relación entre disponibilidad y reclamos?
disponibilidad = df["bicicletas_disponibles"]
reclamos = df["reclamos"]

print("Estaciones con mayor demanda:")
display(demanda_estacion.head(5))

print("Comunas con más reclamos:")
display(reclamos_comuna)

print("Demanda diaria:")
display(demanda_fecha.head())




# ============================================================
# 5. CALCULAR INDICADORES ESTADÍSTICOS
# ============================================================

# Se usa una función externa compartida por otro integrante
indicadores = calcular_indicadores(
    df["prestamos_dia"],
    "indicadores_prestamos_bikecity.joblib"
)

print("Indicadores estadísticos de préstamos diarios:")
print(indicadores)



# ============================================================
# 7. DASHBOARD 2X2 CON DIFERENTES TIPOS DE GRÁFICOS
# ============================================================

plt.figure(figsize=(15, 10))

# Gráfico 1: Barras
# Responde: ¿Qué estaciones presentan mayor demanda?
plt.subplot(2, 2, 1)
demanda_estacion.head(5).plot(kind="bar")
plt.title("1. Estaciones con mayor demanda")
plt.xlabel("Estación")
plt.ylabel("Préstamos promedio")
plt.xticks(rotation=45)

# Gráfico 2: Circular
# Responde: ¿Qué comunas presentan más reclamos?
plt.subplot(2, 2, 2)
reclamos_comuna.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)
plt.title("2. Distribución de reclamos por comuna")
plt.ylabel("")

# Gráfico 3: Línea
# Responde: ¿Cómo evoluciona la demanda en el tiempo?
plt.subplot(2, 2, 3)
demanda_fecha.plot(
    kind="line",
    marker="o"
)
plt.title("3. Evolución diaria de préstamos")
plt.xlabel("Fecha")
plt.ylabel("Total de préstamos")
plt.xticks(rotation=45)

# Gráfico 4: Dispersión
# Responde: ¿Existe relación entre disponibilidad y reclamos?
plt.subplot(2, 2, 4)
plt.scatter(
    disponibilidad,
    reclamos
)
plt.title("4. Disponibilidad vs reclamos")
plt.xlabel("Bicicletas disponibles")
plt.ylabel("Reclamos")

plt.tight_layout()

# Guardar dashboard como imagen para incluirlo en el informe
plt.savefig(
    "dashboard_bikecity_2x2.png",
    dpi=300
)

plt.show()


# ============================================================
# 8. EXPORTAR REPORTES PARA DOS ÁREAS
# ============================================================

# Se usan funciones externas compartidas por el equipo
reporte_operaciones = crear_reporte_operaciones(
    df,
    "reporte_operaciones_bikecity.xlsx"
)

reporte_clientes = crear_reporte_clientes(
    df,
    "reporte_clientes_bikecity.xlsx"
)

print("Reporte de operaciones generado:")
display(reporte_operaciones.head())

print("Reporte de atención al cliente generado:")
display(reporte_clientes.head())


# ============================================================
# 9. INTERPRETACIÓN DE RESULTADOS
# ============================================================

estacion_mayor_demanda = demanda_estacion.index[0]
valor_mayor_demanda = demanda_estacion.iloc[0]

comuna_mas_reclamos = reclamos_comuna.index[0]
valor_mas_reclamos = reclamos_comuna.iloc[0]

print("Interpretación de resultados:")

print(
    f"La estación con mayor demanda promedio es {estacion_mayor_demanda}, "
    f"con {valor_mayor_demanda:.2f} préstamos diarios promedio."
)

print(
    f"La comuna con mayor cantidad de reclamos es {comuna_mas_reclamos}, "
    f"con {valor_mas_reclamos} reclamos acumulados."
)

print(
    "La evolución diaria de préstamos permite identificar fechas con mayor uso del servicio."
)

print(
    "El gráfico de dispersión permite observar si una menor disponibilidad de bicicletas "
    "se relaciona con una mayor cantidad de reclamos."
)


# ============================================================
# 10. DECISIONES RECOMENDADAS
# ============================================================

print("Decisiones recomendadas:")

print("- Redistribuir bicicletas hacia estaciones con mayor demanda promedio.")
print("- Reforzar estaciones ubicadas en comunas con mayor cantidad de reclamos.")
print("- Monitorear los días de mayor demanda para planificar reposición.")
print("- Evaluar estaciones con baja disponibilidad para evitar pérdida de servicio.")
print("- Usar los reportes exportados para apoyar decisiones de operaciones y atención al cliente.")



```
